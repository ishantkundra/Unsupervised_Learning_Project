### 3. Clustering: [15 Marks]

#### A. Apply K-Means clustering for 2 to 10 clusters.

```python
one_hot = pd.get_dummies(data_df['origin'])
one_hot = one_hot.add_prefix('origin_')

data_df = data_df.join(one_hot) 
data_df.head()
```

```python
data_df['hp']=data_df['hp'].replace('?',np.nan)
data_df[data_df['hp']=='?'].count()
```

```python
data_df.fillna(data_df.mean(),inplace=True)
```

```python
data_df['hp'] = data_df['hp'].astype(float)
```

```python
## data_numeric
data_df_numeric = data_df.select_dtypes(include=[np.int64, np.float64]).columns # Choose numeric columns from the dataframe
data_df_numeric
```

```python
from scipy.stats import zscore
```

```python
data_df_numeric=data_df[data_df_numeric].apply(zscore)
data_df_numeric.head()
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

meanDistortions=[]
cluster=range(2,11)

for k in cluster:
    model=KMeans(n_clusters=k)
    model.fit(data_df_numeric)
    prediction=model.predict(data_df_numeric)
    meanDistortions.append(sum(np.min(cdist(data_df_numeric,model.cluster_centers_,'euclidean'),axis=1))/data_df_numeric.shape[0])
```

#### B. Plot a visual and find elbow point. [2 Marks]

```python
plt.plot(cluster,meanDistortions,'bx-')
plt.xlabel('Clusters')
plt.ylabel('Mean Distortions')
plt.title("Select optimum clusters")
```

#### C. On the above visual, highlight which are the possible Elbow points. [1 Marks]

```python

plt.plot(cluster,meanDistortions, 'bx-')

plt.plot([4, 5],[1.49,1.34],'r*', markersize=12, label='Possible Elbow Points')
plt.legend(loc = 'best')
plt.show()
```

#### D. Train a K-means clustering model once again on the optimal number of clusters. [3 Marks]

```python
final_model=KMeans(5)

final_model.fit(data_df_numeric)
prediction=final_model.predict(data_df_numeric)

```

#### E. Add a new feature in the DataFrame which will have labels based upon cluster value. [2 Marks]

```python
data_df["Group"]=prediction
data_df['Group'] = data_df['Group'].astype('category')
data_df.dtypes
```

#### A. Explain pre-requisite/assumptions of PCA. [2 Marks]

1. PCA assumes that the principal component with high variance must be paid attention and the PCs with lower variance are disregarded as noise. Pearson correlation coefficient framework led to the origin of PCA, and there it was assumed first that the axes with high variance would only be turned into principal componentsm

2. All variables should be accessed on the same ratio level of measurement. The most preferred norm is at least 150 observations of the sample set with a ratio measurement of 5:1.

3. Extreme values that deviate from other data points in any dataset, which are also called outliers, should be less. More number of outliers will represent experimental errors and will degrade your ML model/algorithm.

4. The feature set must be correlated and the reduced feature set after applying PCA will represent the original data set but in an effective way with fewer dimensions.

#### B. Explain advantages and limitations of PCA. [3 Marks]

#### Advantages

1. Easy to compute. PCA is based on linear algebra, which is computationally easy to solve by computers.
2. Speeds up other machine learning algorithms. Machine learning algorithms converge faster when trained on principal components instead of the original dataset.
3. Counteracts the issues of high-dimensional data. High-dimensional data causes regression-based algorithms to overfit easily. By using PCA beforehand to lower the dimensions of the training dataset, we prevent the predictive algorithms from overfitting. 

#### Limitations

1. Low interpretability of principal components. Principal components are linear combinations of the features from the original data, but they are not as easy to interpret. For example, it is difficult to tell which are the most important features in the dataset after computing principal components. 

2. The trade-off between information loss and dimensionality reduction. Although dimensionality reduction is useful, it comes at a cost. Information loss is a necessary part of PCA. Balancing the trade-off between dimensionality reduction and information loss is unfortunately a necessary compromise that we have to make when using PCA.
